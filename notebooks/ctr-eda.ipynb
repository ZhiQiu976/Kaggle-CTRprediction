## Data

- Restricted data size to the first one million records.

```python
col_names_train = ['Label'] + \
["I"+str(i) for i in range(1, 14)] + \
['C'+str(i) for i in range(1,27)]

col_names_test = col_names_train[1:]

# # dask
# df_train = dd.read_csv('/kaggle/input/criteo-dataset/dac/train.txt', 
#                        sep='\t', names=col_names_train)

# df_test = dd.read_csv('/kaggle/input/criteo-dataset/dac/test.txt', 
#                        sep='\t', names=col_names_test)
```

```python
# numpy read in: chunksize
# 返回的reader是TextFileReader类型，
# 它指向若干个chunk位置，只在访问数据的时候才真正把数据读入到内存。
# 每个chunk都是dataframe类型的
df_train = pd.read_csv('/kaggle/input/criteo-dataset/dac/train.txt', 
                       sep='\t', names=col_names_train,
                       chunksize=100000) # ten chunks: first 1,000,000

df_test = pd.read_csv('/kaggle/input/criteo-dataset/dac/test.txt', 
                      sep='\t', names=col_names_test,
                      chunksize=100000)
```

```python
%%time

# don't re-run, getting without replacement
df_train_100 = df_train.get_chunk(1000000)
#df_test_100 = df_test.get_chunk(1000000)

# Using the first one million records for analysis
```

```python
df_train_100.shape # one million data, 39 features, 13 numerical, 26 categorical
```

```python
df_train_100 = df_train_100.convert_dtypes()
df_train_100.info()
```

```python
df_train_100.head() # pd.NA (pandas' object to indicate a missing value)
```

## EDA

```python
# missing data percentage for each feature
df_train_100.isnull().sum() / df_train_100.shape[0]
```

```python
# target variable
plt.hist(df_train_100['Label'])
plt.xlabel("Label")
plt.ylabel("Count")
plt.title("Distribution of Labels")
plt.show()
```

```python
# neg. label vs pos. label
df_train_100.Label.value_counts() / df_train_100.shape[0]
```

```python
# Numerical Features
for col in col_names_train[1:14]:
    print('Summary for variable ' + col + ':')
    print(df_train_100[col].describe())
#     plt.hist(df_train_100['I1'][df_train_100['I1'].notna()])
#     plt.show()
    print('-'*60)
```

```python
# pair scatter plot
plt.figure(figsize=(20, 18))
sns.pairplot(data=df_train_100, vars=col_names_train[1:14]) # dropna=True in default
plt.show()
```

```python
# correlation matrix
cor_matrix = df_train_100[col_names_train[1:14]].corr() 
plt.figure(figsize=(12, 10))
sns.heatmap(cor_matrix, annot=True, cmap = 'Reds') 
plt.title('Correlation Matrix for Numeric Variables') 
plt.show()
```

```python
# Categorical Features
for col in col_names_train[14:]:
    print('Number of levels for variable '+col+' is: {}'.format(
    df_train_100[col].nunique()))
    print('Percentage counts for top 20 levels:')
    print(df_train_100[col].value_counts(normalize=True)[:20])
    # plot for 50
    df_train_100[col].value_counts(normalize=True)[:50].plot.bar()
    plt.title("Percentage bar plot for top 50 levels:")
    plt.show()
    print('-'*60)
```
